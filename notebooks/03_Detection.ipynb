# Object Detection on Images
https://github.com/ultralytics/yolov5/issues/36

```python
import torch
import cv2
import os
import pandas as pd
```

```python
img_dir = "./img/"
```

```python
model = torch.hub.load('ultralytics/yolov5', 'yolov5s')
```

```python
imgs = []
for filename in os.listdir(img_dir):
    img = cv2.imread(os.path.join(img_dir, filename))
    if img is not None:
        imgs.append(img)
```

```python
results = model(imgs)
```

```python
type(results.pandas().xyxy[0])
```

```python
frame_count = len(results.pandas())
#df = pd.concat([results.pandas().xyxy[frame].insert(0, 'Frame', frame) for frame in range(frame_count)])
df = pd.DataFrame()
for frame in range(frame_count):
    f = results.pandas().xyxy[frame]
    f.insert(0, "Frame", frame)
    df = pd.concat([df, f])
    
```
