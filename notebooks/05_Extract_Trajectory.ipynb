```python
def get_txy_from_grid(dataframe, obj_id, grid):
    dataframe = dataframe[dataframe["Id"].isin(obj_id)]
    result = pd.DataFrame()
    result["t"] = dataframe["Frame"]
    result["x"] = dataframe["x"] + 0.5 * dataframe["w"]
    result["y"] = dataframe["y"] + 0.5 * dataframe["h"]
    
    return result
```
